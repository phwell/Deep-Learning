```python
# Problema da iris

import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
iris = pd.read_csv('iris.csv')
```

```python
# target é class

iris
```

```python
# Pega os valores da coluna 0 até a coluna 3
entradas = iris.iloc[:, 0:4].values
```

```python
entradas
```

```python
# normalizando entradas
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
entradas = sc.fit_transform(entradas)
```

```python
classe = iris.iloc[:, 4].values
```

```python
classe
```

```python
# Vamos transformar a saída/classe/target
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
classe_codificada = labelEncoder.fit_transform(classe)
```

```python
classe_codificada
```

```python
from keras.utils import np_utils
classe_dummy = np_utils.to_categorical(classe_codificada)
```

```python
classe_dummy.shape
```

```python
classe_dummy
```

```python
def criar_modelo(otimizador, funcao_ativacao, neuronios, dropout, inicializador_kernel):
    classificador = Sequential()
    classificador.add(Dense(units=neuronios, activation=funcao_ativacao,
                            kernel_initializer=inicializador_kernel, input_shape=(4,)))
    classificador.add(Dropout(dropout))
    classificador.add(Dense(units=neuronios, activation=funcao_ativacao, 
                            kernel_initializer=inicializador_kernel))
    classificador.add(Dropout(dropout))
    classificador.add(Dense(units=3, activation='softmax'))
    classificador.compile(optimizer=otimizador, loss='categorical_crossentropy', 
                          metrics=['accuracy'])
    return classificador

classificador_iris = KerasClassifier(build_fn=criar_modelo)

parametros= {'batch_size':[10,30],
             'epochs': [2000,3000],
             'dropout': [0.2, 0.3],
             'otimizador': ['adam', 'sgd'],
             'inicializador_kernel': ['random_uniform', 'normal'],
             'funcao_ativacao': ['relu', 'sigmoid', 'tanh'],
             'neuronios': [4, 8, 16]
            }

grid_search = GridSearchCV(estimator=classificador_iris, param_grid=parametros, cv=2)
grid_search = grid_search.fit(entradas, classe_dummy)
```

```python
# Após a linha acima executar
melhores_parametros = grid_search.best_params_
melhor_precisao = grid_search.best_score_
```
