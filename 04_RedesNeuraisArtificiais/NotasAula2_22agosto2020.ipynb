```python
X_train = temperatura_df['Celsius']
```

```python
y_train = temperatura_df['Fahrenheit']
```

```python
y_train.shape
```

# Criação do modelo usando Keras

```python
import tensorflow as tf

# cria objeto para nosso modelo
modelo_previsao_temperatura = tf.keras.Sequential()

# entrada definido pelo parâmetro input_shape
# saída definida pelo parâmetro units
modelo_previsao_temperatura.add( tf.keras.layers.Dense(units=1 ,input_shape=[1]) )
```

```python
modelo_previsao_temperatura.summary()
```

```python
#https://keras.io/
taxa_aprendizagem = 0.1
modelo_previsao_temperatura.compile(optimizer=tf.keras.optimizers.Adam(taxa_aprendizagem), 
                                    loss='mean_squared_error')
```

# Treinamento

```python
epocas = 500

historico_epocas = modelo_previsao_temperatura.fit(X_train, y_train, epochs=epocas)
```

# Avaliar o modelo

```python
historico_epocas.history.keys()
```

```python
historico_epocas.history['loss']
```

```python
import matplotlib.pyplot as plt
plt.plot(historico_epocas.history['loss'])
plt.title('Evolução Erro durante o treinamento')
plt.xlabel('Épocas')
plt.ylabel('Erro')
plt.legend(['Perda durante o treinamento'])
```
