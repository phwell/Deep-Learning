# Observações

Se quisermos fazer melhor do que ações aleatórias em cada etapa, provavelmente seria bom saber o que nossas ações estão fazendo com o meio ambiente.

A função step do ambiente retorna exatamente o que precisamos. Na verdade, step retorna quatro valores. Esses são:

* observation (object): um objeto específico do ambiente que representa sua observação do ambiente. Por exemplo, dados de pixel de uma câmera, ângulos e velocidades de articulação de um robô ou o estado do tabuleiro em um jogo de tabuleiro.

* reward (float): quantidade de recompensa alcançada pela ação anterior. A escala varia entre os ambientes, mas o objetivo é sempre aumentar sua recompensa total.

* done (boolean): se é hora de redefinir o ambiente novamente. A maioria das tarefas (mas não todas) é dividida em episódios bem definidos e, se for True, indica que o episódio terminou. (Por exemplo, talvez o mastro esteja muito inclinado ou você tenha perdido sua última vida.)

* info (dict):  informações de diagnóstico úteis para depuração. Às vezes, pode ser útil para o aprendizado (por exemplo, pode conter as probabilidades brutas por trás da última mudança de estado do ambiente). No entanto, as avaliações oficiais do seu agente não têm permissão para usar isso para aprendizagem.

Esta é apenas uma implementação do clássico “agent-environment loop”. A cada passo de tempo, o agente escolhe uma ação, e o ambiente retorna uma observação e uma recompensa.

O processo é iniciado chamando reset(), que retorna uma observação inicial. Portanto, uma maneira mais adequada de escrever o código anterior seria respeitar o sinalizador concluído:


Esses ambientes são ótimos para aprender, mas eventualmente você vai querer configurar um agente para resolver um problema personalizado. Para fazer isso, você precisará criar um ambiente personalizado, específico para o domínio do seu problema. 
Primeiro, vamos aprender sobre o que exatamente é um ambiente. Um ambiente contém todas as funcionalidades necessárias para executar um agente e permitir que ele aprenda. Cada ambiente deve implementar a seguinte interface:

No construtor, primeiro definimos o tipo e a forma de nosso action_space, que conterá todas as ações possíveis para um agente realizar no ambiente. Da mesma forma, vamos definir a observation_space, que contém todos os dados do ambiente a serem observados pelo agente.
Nosso método de redefinição será chamado para redefinir periodicamente o ambiente para um estado inicial. Seguem-se várias etapas pelo ambiente, nas quais uma ação será fornecida pelo modelo e deverá ser executada, e a próxima observação retornada. É aqui também que as recompensas são calculadas, mais sobre isso mais tarde.

Finalmente, o método render pode ser chamado periodicamente para imprimir uma representação do ambiente.


```python
!pip install gym_anytrading

import gym
import gym_anytrading
from gym_anytrading.envs import ForexEnv, StocksEnv 
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline  

def my_process_data(env):
    start = env.frame_bound[0] - env.window_size
    end = env.frame_bound[1]
    prices = env.df.loc[:, 'Low'].to_numpy()[start:end]
    signal_features = env.df.loc[:, ['Close', 'Open', 'High', 'Low']].to_numpy()[start:end]
    return prices, signal_features


class MyForexEnv(ForexEnv):
    _process_data = my_process_data


def my_process_data(df, window_size, frame_bound):
    start = frame_bound[0] - window_size
    end = frame_bound[1]
    prices = df.loc[:, 'Low'].to_numpy()[start:end]
    signal_features = df.loc[:, ['Close', 'Open', 'High', 'Low']].to_numpy()[start:end]
    return prices, signal_features


class MyStocksEnv(StocksEnv):
    
    def __init__(self, prices, signal_features, **kwargs):
        self._prices = prices
        self._signal_features = signal_features
        super().__init__(**kwargs)

    def _process_data(self):
        return self._prices, self._signal_features

    
def getAcoesDF(url):
    df1 = pd.read_csv(url)
    df1.dropna(inplace=True)
    #df.drop(['Adj Close'], axis=1, inplace=True)
    df1.to_csv('./acoes.csv')

    df1= pd.read_csv('./acoes.csv')
    df1= df1.sort_values('Date')
    
    return df1

def getProposta(posicao):
    if posicao == 0:
        return 'VENDER'
    
    if posicao == 1:
        return 'COMPRAR'

def getPosition(posicao1, posicao2):
    if posicao1 != posicao2:
        return 'INCONCLUSIVO'
    
    return getProposta(posicao1)

    
def avaliacao(cod):
    url = 'https://query1.finance.yahoo.com/v7/finance/download/'+cod+'.SA?period1=946857600&period2=1586563200&interval=1d&events=history'

    print(cod, url, 'Method 1 (Recommended)')
    df1=getAcoesDF(url)

    #Method 1 (Recommended):
    env = MyForexEnv(df=df1, window_size=12, frame_bound=(12, len(df1)))
    observation = env.reset()
    while True:
        action = env.action_space.sample()
        observation, reward, done, info1 = env.step(action)
        # env.render()
        if done:
            print("info:", info1, 'VENDER=0 and COMPRAR=1')
            break

    plt.cla()
    env.render_all()
    plt.show()

    #Method 2:
    print(cod, url, 'Method 2')
    df1=getAcoesDF(url)
    prices, signal_features = my_process_data(df=df1, window_size=30, frame_bound=(30, len(df1)))
    env = MyStocksEnv(prices, signal_features, df=df1, window_size=30, frame_bound=(30, len(df1)))
    observation = env.reset()
    while True:
        action = env.action_space.sample()
        observation, reward, done, info2 = env.step(action)
        # env.render()
        if done:
            print("info:", info2, 'VENDER=0 and COMPRAR=1')
            break

    plt.cla()
    env.render_all()
    plt.show()    
    
    return (getProposta(info1['position']), getProposta(info2['position']), getPosition(info1['position'], info2['position']))

#codigos_acoes=['PETR4', 'AZUL4', 'ABEV3','MGLU3', 'EMBR3','GOLL4','IRBR3','BBDC4','IGTA3','ALPA4','JBSS3','RENT3', 'RAIL3', 
#               'BBAS3','VALE3','ITUB4','YDUQ3','GNDI3','HAPV3','B3SA3','ELET3','ITUB3','CSNA3','USIM5','CMIG4','LREN3', 'CNTO3']
codigos_acoes=['MGLU3', 'CNTO3']
resultado=[]
for i in range (1,10):
  for cod in codigos_acoes:
    print(cod)
    (metodo1, metodo2, proposta) = avaliacao(cod)
    resultado.append([cod, metodo1, metodo2, proposta])
    df = pd.DataFrame(resultado, columns=['cod', 'metodo1', 'metodo2', 'proposta'])
    print(df)

dffinal=df.groupby(['cod','proposta'])['proposta'].count()
dffinal.to_csv('resultado_acoes.csv')
dffinal
```
