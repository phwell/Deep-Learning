```python
# importar bibliotecas

from keras.models import Sequential 
from keras.layers import Dense, Dropout, LSTM
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# https://keras.io/api/layers/recurrent_layers/lstm/
# [batch, timesteps, feature]
previsores = np.reshape(previsores, (previsores.shape[0], previsores.shape[1], 1))
```

```python
previsores.shape[0]
```

```python
previsores.shape
```

```python
modelo = Sequential()

# numero de celulas de memÃ³ria units = 100
# return_sequences = True indica q vai passar a informacao para frente
# input_shape = (previsores.shape[1], 1) 90,1
modelo.add(LSTM(units = 100, return_sequences = True, input_shape = (previsores.shape[1], 1)))
modelo.add(Dropout(0.3)) # prever overfitting

modelo.add(LSTM(units = 50, return_sequences = True))
modelo.add(Dropout(0.3))

modelo.add(LSTM(units = 50, return_sequences = True))
modelo.add(Dropout(0.3))

modelo.add(LSTM(units = 50))
modelo.add(Dropout(0.3))

# rede densa
modelo.add(Dense(units = 1, activation = 'linear'))
```

```python
# https://keras.io/api/optimizers/rmsprop/
modelo.compile(optimizer = 'rmsprop', loss = 'mean_squared_error', metrics = ['mean_absolute_error'])
```

```python
epocas=10
modelo.fit(previsores, preco_real, epochs = epocas, batch_size = 32)
```

```python
#base_teste = pd.read_csv('petr4_teste.csv')
base_teste = pd.read_csv('PETR4SA_baixado08setembro2020_teste.csv')
preco_real_teste = base_teste.iloc[:, 1:2].values # pegamos todas as linhas e a primeira coluna

base_completa = pd.concat((acoes['Open'], base_teste['Open']), axis = 0) # axis = 0 concatena por coluna

entradas = base_completa[len(base_completa) - len(base_teste) - 90:].values  # retorna as Ãºltimas aÃ§Ãµes que devem ser buscadas
entradas = entradas.reshape(-1, 1) # -1 nao vou trabalhar com linhas 1 para 1 coluna
entradas = normalizador.transform(entradas)

print('len(base_completa) ',len(base_completa), 'len(base_teste) ', len(base_teste), (len(base_completa) - len(base_teste) - 90), base_completa[len(base_completa) - len(base_teste) - 90:].values)

X_teste = []
for i in range(90, 112):  # 90 + 22(teste) = 112
    X_teste.append(entradas[i-90:i, 0])
X_teste = np.array(X_teste)
X_teste = np.reshape(X_teste, (X_teste.shape[0], X_teste.shape[1], 1))
previsoes = modelo.predict(X_teste)
previsoes = normalizador.inverse_transform(previsoes) # desnormalizar

previsoes.mean()
preco_real_teste.mean()
```
